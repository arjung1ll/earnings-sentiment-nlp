```python
def load_transcript(filename):
    """Load a transcript from the data/raw folder."""
    path = os.path.join("..", "data", "raw", filename)
    with open(path, "r", encoding="utf-8") as f:
        return f.read()

def clean_transcript(text):
    """
    Basic cleaning:
    - Remove operator instructions (e.g. 'Please hold...')
    - Remove Q&A section headers
    - Strip extra whitespace
    """
    lines = text.split("\n")
    cleaned = []
    for line in lines:
        line = line.strip()
        if not line:
            continue
        # Skip operator instructions
        if line.lower().startswith(("operator:", "please hold", "ladies and gentlemen")):
            continue
        cleaned.append(line)
    return " ".join(cleaned)

# Test on a transcript
# text = load_transcript("barclays_q4_2023.txt")
# cleaned = clean_transcript(text)
# print(f"Original: {len(text)} chars → Cleaned: {len(cleaned)} chars")
```
